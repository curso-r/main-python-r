```python
import pandas as pd
```

```python
imdb = pd.read_csv("../dados/imdb.csv")
```

```python
diretor_unnest = (imdb.
    filter(["direcao"]).
    assign(direcao2 = lambda x: x.direcao.str.split(",")).
    explode("direcao2")
)
diretor_unnest
```

```python
(diretor_unnest.
    groupby(lambda x: x).
    # groupby("direcao2")
    agg(
        diretor_nest = ("direcao2", lambda x: [i for i in x])
    )
)
```
